```python
df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors = 'coerce')
df['Total Spent'].apply(type).value_counts()
```

```python
#Inferring, Quantity based on Price Per Unit and Total Spent

mask_1 = (
    df['Total Spent'].notna() &
    df['Quantity'].isna() &
    df['Price Per Unit'].notna()
)

df.loc[mask_1, 'Quantity'] = (
    df.loc[mask_1, 'Total Spent'] / df.loc[mask_1, 'Price Per Unit']
)

df['Quantity'].isna().sum()
```

```python
#Inferring Price Per Unit

mask_2 = (
    df['Total Spent'].notna() &
    df['Quantity'].notna() &
    df['Price Per Unit'].isna()
)

df.loc[mask_2, 'Price Per Unit'] = (
    df.loc[mask_2, 'Total Spent'] / df.loc[mask_2, 'Quantity']
)

df['Price Per Unit'].isna().sum()
```

```python
#Inferring Total Spent
#Inferring Price Per Unit

mask_3 = (
    df['Total Spent'].isna() &
    df['Quantity'].notna() &
    df['Price Per Unit'].notna()
)

df.loc[mask_3, 'Total Spent'] = (
    df.loc[mask_3, 'Price Per Unit'] * df.loc[mask_3, 'Quantity']
)

df['Total Spent'].isna().sum()
```

```python
mask_valid = df[['Quantity', 'Price Per Unit', 'Total Spent']].notna().all(axis=1)

tolerance = 0.01
df['Total_Spent_Valid'] = False  # default
df.loc[mask_valid, 'Total_Spent_Valid'] = (
    np.abs(df.loc[mask_valid, 'Quantity'] * df.loc[mask_valid, 'Price Per Unit'] - df.loc[mask_valid, 'Total Spent'])
    < tolerance
)
```

```python
df['Total_Spent_Valid'].value_counts()
df[df['Total_Spent_Valid'] == False]
```
